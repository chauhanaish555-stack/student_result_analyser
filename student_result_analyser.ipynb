importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
     'Name':['preeti','akansha','rahul','rohan','deepak','deepika'],
    'maths':[60,80,78,52,45,82],
    'science':[42,50,99,78,50,50],
    'english':[70,65,25,70,82,36],
    'sst':[80,65,79,81,50,98]
    
}
df=pd.DataFrame(data)
df
```

average marks 

```python
df['average']=(df['maths']+df['science']+df['english']+df['sst'])/3
df
```

```python
topper=df[df['average']==df['average'].max()]
print("topper: ")
print(topper)
failures=df[(df['maths']<50) | (df['science']<50) | (df['english']<50) | (df['sst']<50)]
print("\nstudnets who need attention:")
print(failures)                                                    
```

```python
plt.figure(figsize=(10,5))#size of the graph
plt.bar(df['Name'],df['average'],color='lightpink')#plotting of bar graph
plt.title('student average marks')
plt.xlabel('student')
plt.ylabel('marks of the students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("="*40)
print("            STUDENT RESULT ANALYSIS")
print("="*40)
print(f"\ntotal students: {len(df)}")#length of students
print(f"\nclass avg: {df['average'].mean():.2f}")#round off to 2 decimal places
print(f"\ntoppe()s: {list(topper['Name'].values)}")
print(f"\nstudents needing attention: {list(failures['Name'].values)}")
print("\nSubject Wise Class average:")
print(f"  Maths   : {df['maths'].mean():.2f}")
print(f"  Science : {df['science'].mean():.2f}")
print(f"  English : {df['english'].mean():.2f}")
print("="*40)
```
